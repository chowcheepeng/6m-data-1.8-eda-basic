# Categoricals
Categorical data is a type of data that can take on a limited, fixed number of possible values. These values are often referred to as categories or levels.

The `pandas` library in Python provides a powerful data structure called `Categorical` to handle categorical data efficiently. The `Categorical` type can be used to represent categorical variables, which can be either nominal (unordered) or ordinal (ordered).

```python
import pandas as pd
import numpy as np
```

## Creating Sample Data

Let's create a simple dataset representing t-shirt orders with different sizes. We'll use this to explore how pandas handles categorical data.

```python
df = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5, 6, 7],
    "size":     ["S", "M", "L", "M", "S", "L", "XL"],
    "price":    [10, 15, 20, 15, 10, 20, 25]
})

df
```

## Understanding Data Types

By default, pandas stores text columns like "size" as `object` dtype (essentially strings). Let's verify this:

```python
# Check size dtype
df["size"].dtype
```

As we can see, the dtype is `dtype('O')` where 'O' stands for "object" (essentially strings). Now let's convert it to a categorical type, which is more memory-efficient and allows for better analysis.

```python
# Convert size to category dtype
df["size"] = df["size"].astype("category")
df["size"].dtype
```

## Exploring Categories

Now that we've converted to categorical, we can see what unique categories exist:

```python
# Check categories
df["size"].cat.categories
```

### Memory Efficiency: Category Codes

Behind the scenes, pandas stores categorical data as integer codes rather than repeating strings. This saves memory, especially with large datasets:

```python
# Check category codes
df["size"].cat.codes

# Pandas is storing integers (0, 1, 2) instead of the strings "L", "M", "S". The display still shows the strings - that's just pandas decoding them for you.
```

## The Problem with Unordered Categories

Without specifying an order, categorical comparisons will **raise an error**. The categories are just different labels, not values that can be compared:

*(Note: The code below will raise a TypeError - that's the point! It shows why we need ordered categoricals)*

```python
# This will raise a TypeError because unordered categories cannot be compared
# Uncomment to try:
df[df["size"] >= "M"]
```

```python
# Try sorting
df.sort_values("size")
```

Sorting also happens alphabetically by default, not by the actual size order we might expect. "L" comes before "M" alphabetically, but we know that Medium is smaller than Large in terms of clothing sizes.

## Ordered Categoricals: The Solution

To fix this, we can create an **ordered categorical** by specifying the categories in the correct order. Now pandas knows that S < M < L < XL:

```python
# Use case - add a meaningful order to the categories
df["size"] = pd.Categorical(df["size"], categories=[
                            "S", "M", "L", "XL"], ordered=True)
```

```python
df.sort_values("size", ascending=False)
```

Now sorting works correctly! The sizes are ordered from smallest to largest (or vice versa with `ascending=False`):

```python
df[df["size"] >= "L"]
```

## Summary

**Key Takeaways:**

1. **Categorical dtype** saves memory by storing categories as integers internally
2. **Unordered categories** are just labels - you can't compare or sort them meaningfully
3. **Ordered categories** allow you to:
   - Sort data in meaningful order
   - Use comparison operators (>, <, >=, <=)
   - Filter data based on category ranking

**When to use ordered categoricals:**
- T-shirt sizes (XS, S, M, L, XL, XXL)
- Education levels (High School, Bachelor's, Master's, PhD)
- Satisfaction ratings (Poor, Fair, Good, Excellent)
- Any data where categories have a natural order
